```python
class BinaryHeap:
  def _swap(self, i, j):
    self._container[i], self._container[j] = self._container[j], self._container[i]

  def _parent(self, i):
    return (i - 1) // 2
  
  def _left(self, i):
    return 2 * i + 1

  def _right(self, i):
    return 2 * i + 2

  def _has_left(self, i):
    return self._left(i) < self._size
  
  def _has_right(self, i):
    return self._right(i) < self._size
  
  def _upheap(self, i):
    parent = self._parent(i)
    if i > 0 and self._container[i][0] < self._container[parent][0]:
      self._swap(i, parent)
      self._upheap(parent)
  
  def _downheap(self, i):
    if self._has_left(i):
      left = self._left(i)
      small_child = left
      if self._has_right(i):
        right = self._right(i)
        if self._container[right][0] < self._container[left][0]:
          small_child = right
      if self._container[small_child][0] < self._container[i][0]:
        self._swap(i, small_child)
        self._downheap(small_child)
  
  def __init__(self, arr=[]):
    self._container = arr
    self._size = len(arr)
    if(self._size > 1):
      self.build_heap(arr)
  
  def push(self, est:tuple):
    self._container.append(est)
    self._size += 1
    self._upheap(self._size - 1)

  def extract_min(self):
    if self._size == 0:
      raise IndexError('Empty Heap!')
    self._swap(0, self._size - 1)
    return_value = self._container.pop()
    self._size -= 1
    self._downheap(0)
    return return_value
  
  def build_heap(self, alist:list):
    i = len(alist) // 2
    self._size = len(alist)
    self._container = alist[:]
    while (i >= 0):
        self._downheap(i)
        i = i - 1
  
  def empty(self)->bool:
    return self._size == 0
  
  def __len__(self):
    return self._size
  
  def __str__(self):
    return str(self._container)
  
  def __repr__(self):
    return str(self._container)

  
  
  
```

```python
def distancia_para_tempo_em_minutos(distancia_km):
  velocidade_km_por_h = 30
  tempo_hrs = distancia_km / velocidade_km_por_h
  tempo_minutos = tempo_hrs * 60
  return tempo_minutos
```
